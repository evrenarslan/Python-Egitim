## Matplotlib Temel Kullanımı

Magic function ***inline*** kullanımı gönderilen fonksiyonun çıktısının kodun çıktısının içinde olmasnı sağlar. Bunun alternatifi olan ***notebook*** ile ise yaratılan objenin çıktısının koddan bağımsız olarak yaratılmasını ve dolayısı ile sonraki kodlarda da aynı objenin kullanımında çıktının etkilenmesini sağlar.

```python
#%matplotlib inline
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
plt.plot(5,5,'o')
plt.show()
```

```python
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Grafk Örneği")
```

## Matplotlib ile yukarıda hazırladığımız verileri görsel hale getirelim.

```python
%matplotlib inline

import matplotlib as mlp
import matplotlib.pyplot as plt
```

matplotlib'in yüklenip yüklenmediğini kontrol edelim.

```python
print("Matplotlib versiyon :",mlp.__version__)
```

```python
print(plt.style.available) # Matplotlib için uygulanabilir stillerin listesini bu şekilde görebilirsiniz.
#İstenirse belirli bir stil uygulanabilir.
mlp.style.use(["ggplot"])
```

Pandas içinde bir dataframe'den grafik çizdirmek için <code>df.plot()</code> kullanılabilir.

Aşağıdaki örnekte Haiti'den Kanada'ya göçmüş olan göçmenlerin sayısının yıllar içerisindeki değişimine bakmak için line grafik çizdireceğim.

```python
haiti=df_can.loc["Haiti",years] # Sadece yıllar boyunca göçmen sayısını alacağım.
haiti.head(3)
```

```python
haiti.plot()
```

```python
haiti.index = haiti.index.map(int) # sene bilgisini integer olarak değiştiriyoruz.
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show() # Yaptığımız değişikliklerin görüntülenmesini istediğim için .show() metodunu kullanıyorum.
```

2010 yılındaki bu yükselişin nedenini google üzerinden araştılır ise https://en.wikipedia.org/wiki/2010_Haiti_earthquake büyük bir deprem olduğu görülecektir.

```python
# Grafik üzerine yukarıdaki açıklamayı ekleyelim.
haiti.index = haiti.index.map(int) # sene bilgisini integer olarak değiştiriyoruz.
haiti.plot(kind='line')

plt.title("Haiti'den Gelen Göçmenler")
plt.ylabel('Göçmen sayısı')
plt.xlabel('Yıllar')

plt.text(2000, 6000, '2010 Depremi') # Grafik üzerine not düşülmesini sağlıyorum. x=2000 ve y=6000 koordinatlarına ekliyorum.

plt.show() # Yaptığımız değişikliklerin görüntülenmesini istediğim için .show() metodunu kullanıyorum.
```

China ve India'dan gelmiş göçmenlerin grafiğini çizdirelim.

```python
df_CI=df_can.loc[["India","China"],years]
df_CI.head()
```

```python
df_CI.plot(kind="line")
```

Grafik hiçde anlamlı gözükmüyor. Fakat traspose edersek doğru gözükecektir.

```python
df_CI=df_CI.transpose()
df_CI.head()
```

```python
df_CI.plot(kind="line")
```

Grafik şu an anlamlı gözüküyor. Yine yukarıda yaptığımız gibi grafiğin verisi üzerinde yıl verisini int yapalım. Grafik üzerine tanımlayıcı verileri ekleyelim. 

```python
df_CI.index=df_CI.index.map(int)
df_CI.plot(kind="line")

plt.title("Çin ve Hindistan'dan Gelen Göçmenler")
plt.ylabel("Göçmen sayısı")
plt.xlabel("Yıllar")

plt.show()
```

***En fazla 5 göçmen gelen ülkenin grafiğini çizdirelim.***

```python
df_can.sort_values(by="Total",ascending=False,axis=0,inplace=True)
```

```python
df_top5=df_can.head(5)
```

```python
df_top5=df_top5[years].transpose()
```

```python
df_top5.head()
```

```python
df_top5.index=df_top5.index.map(int)

df_top5.plot(kind='line', figsize=(14, 8))

plt.title("En Fazla Göçmen Gelen 5 Ülke'nin Trendi")
plt.ylabel('Göçmen sayısı')
plt.xlabel('Yıllar')

plt.show()
```
