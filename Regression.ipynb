```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
path='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv'
```

Öncelikle hangi alanlar arasında korelasyon olduğunu anlamaya çalışacağız. Aşağıdaki tablodan price ile en yüksek korelasyonun engine-size arasında olduğunu görüyoruz.

```python
df.corr()
```

***Tek boyutlu linear regression***

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
X=df[["engine-size"]] # Bunu df olarak alacağım.
```

```python
Y=df["price"] # Bunu ise bir seri olarak alıyorum
```

```python
X
```

```python
Y
```

```python
lm.fit(X,Y)
```

```python
sabit=lm.intercept_
```

```python
egim=lm.coef_
```

```python
sabit
```

```python
egim
```

```python
lm.predict(X)
```

```python
#bulduğumuz değerler ile prediction
x=130
y=sabit+egim*x
y[0]
```

```python
df[df["engine-size"]==130][["engine-size","price"]]
#Gerçek değer
```

Görüldüğü gibi korelasyon yüksek olsa bile tam doğru sonuç diye bir şey yok. Olası en yakın sonucu tahmin etmeye çalışıyoruz. Şimdi parametre sayısını arttırarak regresyon yapalım.  curb-weight horsepower width değerlerini de alalım

```python
X=df[["engine-size","curb-weight","horsepower","width"]] 
```

```python
lm.fit(X,Y)
```

```python
sabit=lm.intercept_
egim=lm.coef_
```

```python
sabit
```

```python
egim
```

```python
df[df["engine-size"]==130][["engine-size","price","curb-weight","horsepower","width"]]
#Gerçek değer
```

```python
#bulduğumuz değerler ile prediction
x=130
z=2548
t=111
p=0.890278
y=sabit+egim[0]*x+egim[1]*z+egim[2]*t+egim[3]*p
y
```
