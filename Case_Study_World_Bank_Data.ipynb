***zip()*** foksiyonu

```python
numbers=[1,2,3]
letters=["a","b","c"]
```

```python
type(numbers)
```

```python
type(letters)
```

```python
zipped=zip(numbers,letters)
```

```python
type(zipped)
```

```python
print(zipped) # Direkt olarak değerleri alamıyoruz. Bunun için ya bir iterasyona sokmak yada liste haline getirmek lazım.
```

```python
zip_list=list(zipped)
```

```python
print(zip_list)
```

```python
# Burada iki tane set objesi tanımladık. Zip ile iki set  birleştirildiğinde normalde sıralı olarak veriler verilmesed zip
# operasyonu ile birleştirildiğinde sıralı hale gelir.
s1={2,3,1}
s2={"b","a","c"} 
```

```python
type(s1)
s1
```

```python
type(s2)
s2
```

```python
zip_set=zip(s1,s2)
```

```python
print(list(zip_set)) # Çıktı sıralı bir şekilde olacaktır. 
```

```python
# Değişik uzunlukta iterable verilirse en kısa olanının element sayısı kadar zip objesi yaratır. Gerisini ignore eder.
list(zip(range(5),range(100)))
```

```python
# Eğer en uzun olana göre zip objesini oluşturup diğerlerini boş bırakmak istiyor isek. 
# zip yerine itertools.zip_longest() kullanabiliriz.
from itertools import zip_longest

numbers = [1, 2, 3]
letters = ['a', 'b', 'c']
longest = range(5)
zipped=zip_longest(numbers,letters,longest,fillvalue="?")
list(zipped)
```
