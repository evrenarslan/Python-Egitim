#  İnternet üzerinden veri çekmek


**İnternet üzerinden dosya indirmek**

```python
from urllib.request import urlretrieve # Bununla internet üzerinden veri çekme işini yapabiliriz.
import pandas as pd
```

```python
url="https://s3.amazonaws.com/assets.datacamp.com/production/course_1606/datasets/winequality-red.csv"
```

```python
urlretrieve(url,"winequality-red.csv") # local dizine verilen isim ile kaydeder.
```

```python
df=pd.read_csv("winequality-red.csv",sep=";") # İndirilen dosyayı dataframe içine yüklüyoruz.
```

```python
print(df.head())
```

**Bir web sayfasını yüklemek**

```python
from urllib.request import urlopen,Request
```

```python
url="https://tr.wikipedia.org/wiki/Anasayfa"
```

```python
request=Request(url) # Request objesi yarattık
response=urlopen(request) # Bu request objesini açıp response objesine yükledik.
html=response.read() # html objesi içine response objesini okuyup doldururuz.
print(html)
response.close() # response objesini kapatırız. 
```

***Beautiful Soup***

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://tr.wikipedia.org/wiki/Anasayfa"
```

```python
r=requests.get(url)
```

```python
html_doc=r.text
```

```python
soup=BeautifulSoup(html_doc)
```

```python
print(soup.prettify())
```

```python
print(soup.title) # html sayfasındaki başlık kısmını getirir.
```

```python
print(soup.get_text()) # html taglarinin dışındaki test datayı getirir.
```

```python
a_tags=soup.find_all("a")

for link in a_tags:
    print(link.get("href"))
```
