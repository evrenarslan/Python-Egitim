# Rastgele sayı üretmek 

Herhangi bir nedenle rastgele sayı üretmek için numpy kullanabiliriz. Bunu numpy.random.rand() ile yapabiliriz. İlk olarak burada üretilen numaraların gerçek rastgele değerler olmadığını pseduo numaralar yani bir matematiksel model ile üretildiğini unutmamak gerekiyor. Bu şekilde ürettiğimiz numaraların her seferinde aynı değerleri ürettiğini sağlamak istiyor isek numpy.random.seed(x) fonksiyonunu kullanabiliriz. Eğer buradaki x değeri yerine aynı değeri verir isek her seferinde aynı değerleri üzereten ve rastgele görünen (gerçekte rastgele değil) sayı değerleri üretebiliriz.

```python
import numpy as np

i =0
while i<10 :
    i=i+1
    print(np.random.rand()) # seed ile bir başlanğıç değeri vermediğim için her seferinde farklı bir sayı dizisi elde ediyorum.
```

```python
import numpy as np

np.random.seed(19) # Bu değeri set ettiğim için her çağırmamda aynı sayı serisi oluşuyor.
i=0
while i<10:
    i=i+1
    print(np.random.rand())
```

***rand() ile rastgele float sayı elde ederim. Eğer randint() kullanır isem integer bir değer elde ederim. randint(x,y) ile x'den başlayan ama y'yi içermeyen bir sayı aralığında integer sayılar elde ederim***

```python
import numpy as np

i=0
while i<10:
    i=i+1
    print(np.random.randint(1,49))
```

***Random Walk: Burada rastgele yazı tura atarak 50 adımlık bir yolu kaç denemede bitirebileceğimizi hesaplatacağım***

```python
import numpy as np

yol=50
mesafe=[]
i=0
while yol!=0:
    i=i+1
    para=np.random.randint(0,2)
    if para==0:
        yol=yol+1
    else:
        yol=yol-1
    mesafe.append(yol)
    print(str(i)+ ". deneme: " + str(yol))
```

```python
# Bir değeri değiştirir iken belirli bir değerden küçük olmamasını istiyorsak max() fonksiyonu kullanılabilir

for i in range(100):
    print(str(i)+". deneme: "+ str(max(30,50-i))) # Bu şekilde tanımlayarak 30'dan daha küçük bir değer almasını engelledik.
```

***Random Walk ve Matplot***

```python
import numpy as np

np.random.seed(19)

random_walk=[0]

for x in range(100):
    step=random_walk[-1]
    dice=np.random.randint(1,7)
    
    if dice<=2:
        step=max(0,step-1)
    elif dice<=5:
        step=step+1
    else:
        step=step+np.random.randint(1,7)
    random_walk.append(step)

import matplotlib.pyplot as plt

plt.plot(random_walk)
plt.show()
```
